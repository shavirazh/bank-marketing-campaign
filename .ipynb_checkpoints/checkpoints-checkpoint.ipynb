## Distribusi kolom kategorik berdasarkan kolom targetnya

```python
df.head()
```

```python
cols = ['job','marital','education','default','housing','loan','contact','poutcome']

fig, ax = plt.subplots(4, 2, figsize=(12,18))
ax = ax.reshape(-1)

for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, hue='y', ax=ax[i])

fig.tight_layout(pad=2.0)
plt.title('Distribusi kolom kategorik')
```

## Korelasi kolom numerik

```python
mapping = {"yes":1,
          "no":0}

df['y'] = df['y'].map(mapping)
```

```python
df.head()
```

```python
plt.figure(figsize=(15, 10))
ax = sns.heatmap(df.corr(), annot=True, cmap="PuRd")
```

## Distribusi kolom numerik

```python
df.head()
```

```python
cols = ['age','duration','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']

fig, ax = plt.subplots(5, 2, figsize=(12,18))
ax = ax.reshape(-1)

for i, col in enumerate(cols):
    sns.histplot(data=df, x=col, hue='y', multiple='stack', ax=ax[i])

fig.tight_layout(pad=2.0)
plt.title('Histogram Kolom Numerik')
```

```python
num_var = ["age","duration","campaign", "pdays", "previous","emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed"]

for i in num_var:
    sns.boxplot(x = 'y',y = i,data = df)
    plt.show();
```
